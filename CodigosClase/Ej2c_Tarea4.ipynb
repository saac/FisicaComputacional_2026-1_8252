```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pyl
```

```python

def IntSimpBessel(a,b,N,f,M,X):

    h = (b-a)/N
    
    if (N %2 == 0):
        
        I = (f(a,M,X) + f(b,M,X))        
        I = I + 4*f(a+(N-1)*h,M,X)
    
        for i in np.arange(1,N/2,1):
            I = I + 4*f(a + (2*i-1)*h,M,X)
            I = I + 2*f(a + 2*i*h,M,X)
    
        I = (h/3)*I

    else:
        print("N debe ser par")

    return I


def f(theta,M,X):
    return np.cos(M*theta - X*np.sin(theta))

def J(m,x):
    S = IntSimpBessel(0,np.pi,1000,f,m,x)   
    return (1/np.pi)*S
    

def I(r,lamb,J,m):
    k = (2*np.pi)/(lamb)
    # r = X/k
    I = (J(m,r*k)/k*r)**2
    return I
```
