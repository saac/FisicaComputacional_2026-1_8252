```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from ipywidgets import *
%matplotlib widget
```

```python
def Malthus(r,P):
    return r*P 
```

```python
r = 3
dt = 0.1

p0 = 0.5
t0 = 0
```

```python
P1 = p0 + Malthus(r,p0)*dt
P1
```

```python
N = 100
p = p0
t = t0

P = [p0]
T = [t0]

for i in range(N):
    p = p + Malthus(r,p)*dt
    t = t + dt

    P.append(p)
    T.append(t)
    # print(t,P)
```

```python
fig, ax = plt.subplots(figsize=((12,8)))
plt.plot(T,P)
```

```python
r = 3

dt = 0.001
p0 = 0.5
t0 = 0
```

```python
p = p0
t = t0
tf = 100

P = [p0]
T = [t0]

while t<=tf:
    p = p + Malthus(r,p)*dt
    t = t + dt

    P.append(p)
    T.append(t)
    # print(t,P)
```

```python
fig, ax = plt.subplots(figsize=((12,8)))
plt.plot(T,P)
# plt.plot(T,P,'r.')
```
