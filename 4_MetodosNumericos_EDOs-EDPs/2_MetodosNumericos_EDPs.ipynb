## 4.3.1. Ecuaciones de **Laplace** y **Poisson**: el _**método de diferencias finitas**_.

### Para poder encontrar una aproximación numerica para las ecuaciones de _Laplace_ y _Poisson_,  primero expandamos el Laplaciano a partir de su definición:

### $\nabla^2 = \nabla \cdot \nabla$

### Por lo que para una función $\phi(x,y)$ dada, tenemos que:

<!-- ### $\nabla^2 \phi = \nabla \cdot \nabla \phi = \dfrac{\partial^2 \phi}{\partial x^2} + \dfrac{\partial^2 \phi}{\partial y^2}$ -->
### $\nabla^2 \phi = \dfrac{\partial^2 \phi}{\partial x^2} + \dfrac{\partial^2 \phi}{\partial y^2}$

### Ahora podemos reescribir cada uno de los terminos de la suma de la siguiente manera:

### $\displaystyle \dfrac{\partial^2 \phi}{\partial x^2} = \dfrac{\partial}{\partial x} \left( \dfrac{\partial \phi(x,y) }{\partial x} \right)$

### $\displaystyle \dfrac{\partial^2 \phi}{\partial y^2} = \dfrac{\partial}{\partial y} \left( \dfrac{\partial \phi(x,y) }{\partial y} \right)$

### Realmente, cada uno de los terminos de la suma implican derivadas parciales, las cuales ya habíamos aproximado numericamente con anterioridad. Por lo que recordando el método de **_"diferencias centrales"_**, podemos reescribir ambos terminos como:

### $\displaystyle \dfrac{\partial^2 \phi}{\partial x^2} = \dfrac{\partial}{\partial x} \left( \dfrac{\partial \phi }{\partial x} \right) = \dfrac{\partial}{\partial x} \left( \dfrac{\phi(x+\frac{h}{2},y)-\phi(x-\frac{h}{2},y)}{h}\right)$

### $\displaystyle \dfrac{\partial^2 \phi}{\partial y^2} = \dfrac{\partial}{\partial y} \left( \dfrac{\partial \phi }{\partial y} \right) = \dfrac{\partial}{\partial y} \left( \dfrac{\phi(x,y+\frac{h}{2})-\phi(x,y-\frac{h}{2})}{h}\right)$


### Que de igual manera, también ya habíamos aproximado numericamente la segunda derivada de una función. Por lo que:

### $\displaystyle \dfrac{\partial^2 \phi}{\partial x^2} = \dfrac{\partial}{\partial x} \left( \dfrac{\phi(x+\frac{h}{2},y)-\phi(x-\frac{h}{2},y)}{h}\right) = \dfrac{\phi(x+h,y) - 2\phi(x,y) + \phi(x-h,y)}{h^2}$

### $\displaystyle \dfrac{\partial^2 \phi}{\partial y^2} = \dfrac{\partial}{\partial y} \left( \dfrac{\phi(x,y+\frac{h}{2})-\phi(x,y-\frac{h}{2})}{h}\right) = \dfrac{\phi(x,y+h) - 2\phi(x,y) + \phi(x,y-h)}{h^2}$


### Así entonces, al sumar ambos terminos tenemos que:

<h3 align="center">$\nabla^2 \phi = \dfrac{\partial^2 \phi}{\partial x^2} + \dfrac{\partial^2 \phi}{\partial y^2} = \dfrac{\phi(x+h,y) - 2\phi(x,y) + \phi(x-h,y)}{h^2} + \dfrac{\phi(x,y+h) - 2\phi(x,y) + \phi(x,y-h)}{h^2}$</h3>

### Por lo que, una aproximación numerica para el Laplaciano de $\phi(x,y)$ sería:
$\require{color}$

<h2 align="center">$ \color{red} \boxed{ \color{black} \nabla^2 \phi = \displaystyle \dfrac{\phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - 4 \phi(x,y)}{h^2} }$</h2>


### 4.3.1.1. **Ecuación de Laplace**.

### Dado lo anterior, ahora nos es posible abordar la ecuación de _Laplace_:

<h2 align="center">$ \nabla^2 \phi = 0$</h2>

### $\;$

### Substituyendo nuestra aproximación numérica del Laplaciano tenemos que:

<h3 align="center">$ \displaystyle \dfrac{\phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - 4 \phi(x,y)}{h^2} = 0 $</h3>

<h3 align="center">$\Rightarrow \displaystyle \phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) = 4 \phi(x,y)$</h3>

#### $\;$

### Esto último resulta bastante conveniente, pues nos proporciona una expresión explicita de $\phi(x,y)$ a partir de la _ecuación de Laplace_:

<h2 align="center">$\phi(x,y) = \frac{1}{4} \displaystyle \left[ \phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) \right]$</h2>

### a partir de la cual, podemos usar el **_método de relajación_** que vimos anteriormente. Es decir, podemos convetir dicha expresión en una ecuación para iterar:

<h2 align="center">$ \Rightarrow \color{blue} \boxed{ \color{black} \phi_n(x,y) = \frac{1}{4} \displaystyle \left[ \phi_{n-1}(x+h,y) + \phi_{n-1}(x-h,y)+ \phi_{n-1}(x,y+h) + \phi_{n-1}(x,y-h) \right] }$</h2>

### Este método en general se conoce como **_"Método de Diferencias Finitas"_**


### Sólo tendríamos que saber quién es $\phi_0$ para poder comenzar:
## $ \phi_0(x,y) = $ ???

### y así calcular las siguientes iteraciones:
## $ \phi_1(x,y) = \frac{1}{4} \displaystyle \left[ \phi_0(x+h,y) + \phi_0(x-h,y)+ \phi_0(x,y+h) + \phi_0(x,y-h) \right]$
### $ \vdots$
## $\phi_n(x,y) = \frac{1}{4} \displaystyle \left[ \phi_{n-1}(x+h,y) + \phi_{n-1}(x-h,y)+ \phi_{n-1}(x,y+h) + \phi_{n-1}(x,y-h) \right]$

### $\;$

### Para conocer $\phi_0$ debemos usar las _**condiciones de Frontera**_

## $\phi_n(0,y) = \phi_{n-1}(0,y) = \phi_0(0,y)$
## $\phi_n(N,y) = \phi_{n-1}(N,y) = \phi_0(N,y)$

## $\phi_n(x,0) = \phi_{n-1}(x,0) = \phi_0(x,0)$
## $\phi_n(x,N) = \phi_{n-1}(x,N) = \phi_0(x,N)$

## **Usando el _"método de relajación"_ considerando _minimización de error_**.

## Ahora, hagamos el mismo calculo (malla de $100 \times 100$), pero ahora iteraremos a partir de **minimizar el error**, como lo hicimos anteriormente cuando abordamos por primera vez el _método de relajación_. Para esto usaremos un ciclo `while` y definiremos dos arreglos para ir guardando la información, `PHIn` y `PHIn_1` para almacenar la información de la iteraión actual $n$ y la de la anterior $n-1$ y que intercambiaremos al final de cada iteración, para poder refrescar el calculo.

```python
# Método de relajación considerando minimización de error

V = 1.25
N = 100

exactitud = 1e-3
error = 1.0

PHIn = np.zeros((N+1,N+1))
PHIn[0,:] = V # condicion de frontera  (formalmente deberia ser PHIn[:,N] = V )

PHIn_1 = np.copy(PHIn)

while error > exactitud:  # hasta minimizar el error!

    for i in range(N+1):  # recorremos todos los indices i de uno en uno (al aumentar el indice, nos movemos al siguiente punto x+h)
        for j in range(N+1):  # recorremos todos los indices j de uno en uno  (al aumentar el indice, nos movemos al siguiente punto y+h)
            
            if i == 0 or i == N or j == 0 or j == N:
                PHIn[i,j] = PHIn_1[i,j]  # aqui no cambia, pues son las condiciones de frontera
            else:
                PHIn[i,j] = (PHIn_1[i+1,j] + PHIn_1[i-1,j] + PHIn_1[i,j+1] + PHIn_1[i,j-1])/4

    error = np.max(np.abs(PHIn_1 - PHIn))
    PHIn_1,PHIn = PHIn,PHIn_1
          
print(PHIn)
fig, ax = plt.subplots(figsize=((12,8)))
plb.imshow(PHIn)
```

# **En resumen**:

## Para usar el método de _**"Diferencias Finitas"**_ para resolver la **Ecuación de Laplace**:
<h1 align="center">$\nabla^2 \phi = 0$</h1> 

## 1) Usamos _"diferencias centrales"_ para aproximar las segundas derivadas parciales y así poder obtener una aproximación númerica del Laplaciano:

<h2 align="center">$ \color{red} \boxed{ \color{black} \nabla^2 \phi = \displaystyle \dfrac{\phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - 4 \phi(x,y)}{h^2} }$</h2>


## 2) Substituimos dicha aproximación numerica en la ecuación de Laplace $\nabla^2 \phi = 0$; y despejamos para obtener una expresión (ecuación tracendental) para $\phi(x,y)$:

<h2 align="center">$ \color{blue} \boxed{ \color{black} \phi(x,y) = \frac{1}{4} \displaystyle \left[ \phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) \right] }$</h2>

## 3) Usando el "método de relajación" itereamos la expresión tracendental para $\phi(x,y)$ en todos los puntos $(x,y)$ de una malla, hasta que converga a la solución.

<h2 align="center">$ \phi_n(x,y) = \frac{1}{4} \displaystyle \left[ \phi_{n-1}(x+h,y) + \phi_{n-1}(x-h,y)+ \phi_{n-1}(x,y+h) + \phi_{n-1}(x,y-h) \right]$</h2>

## 4) En la implementación hay que recorrer todos y cada uno, de los puntos de la malla definida (de forma secuencial), para aproximar numericamente los valores de $\phi_n(x,y)$, en cada iteración, partiendo dede $\phi_0$. Asimismo, para tener los "valores semilla" para cada puntos de $\phi_0(x,y)$ (que en principio pueden ser arbitrarios), hay que usar las _**condiciones de Frontera**_:

## $\phi_n(0,y) = \phi_{n-1}(0,y) = \phi_0(0,y)$
## $\phi_n(N,y) = \phi_{n-1}(N,y) = \phi_0(N,y)$

## $\phi_n(x,0) = \phi_{n-1}(x,0) = \phi_0(x,0)$
## $\phi_n(x,N) = \phi_{n-1}(x,N) = \phi_0(x,N)$

## que siempre tendran los mismos valores independientemente de la iteración $n$.


## 4.3.1.2. **Ecuación de Poisson**.

### Ahora para abordar la ecuación de Poisson con el **_"Método de Diferencias Finitas"_**:
<h2 align="center">$ \nabla^2 \phi = f(x,y)$</h2>

### haremos algo muy similar (practicamente lo mismo a lo hecho para la ecuación de Laplace).

#### $\;$

### **1)** Substituiremos nuestra aproximación numérica del Laplaciano:

<h3 align="center">$ \color{red} \boxed{ \color{black} \nabla^2 \phi = \displaystyle \dfrac{\phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - 4 \phi(x,y)}{h^2} }$</h3>

### en la ecuación, por lo que tendremos que:

<h2 align="center">$\displaystyle \dfrac{\phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - 4 \phi(x,y)}{h^2} = f(x,y)$</h2>

### Reordenando terminos tenemos que:

<h2 align="center">$ \Rightarrow \; \displaystyle \phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - {h^2}f(x,y) = 4 \phi(x,y) $</h2>

<h2 align="center">$ \therefore \; \phi(x,y) = \displaystyle \frac{1}{4}  \left[ \phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) \right] - \frac{h^2}{4}f(x,y) $</h2>

### con lo que al igual que en el caso anterior (de Laplace), obtenemos una expresión (tracendental) para $\phi(x,y)$ a partir de la cual, podemos usar el _método de relajación_ para aproximar la solución. Es decir, podemos convetir dicha expresión en una ecuación para iterar.

### **2)** Usaremos el _método de relajación_ para iterar la expresión tracendental de $\phi(x,y)$ en todos y cada uno de los puntos $(x,y)$ de una malla (de forma secuencial) y usando las _**condiciones de Frontera**_, para aproximar numericamente los valores de $\phi_n(x,y)$, en cada iteración, partiendo dede $\phi_0$, hasta que converga a la solución ($\phi_n$), al igual que como lo hicimos como la _ecuación de Laplace._

<h2 align="center">$ \color{blue} \boxed{ \color{black} \phi_n(x,y) = \displaystyle \frac{1}{4}  \left[ \phi_{n-1}(x+h,y) + \phi_{n-1}(x-h,y)+ \phi_{n-1}(x,y+h) + \phi_{n-1}(x,y-h) \right] - \frac{h^2}{4}f(x,y) }$</h2>

### notese que el último término que contiene a $f(x,y)$ (que hace la diferencia con la Ec. de Laplace), no se itera, pues está difinido a priori y es un valor dado para cada punto $(x,y)$, jugando un rol similar al de las condiciones de frontera (vistos en el caso anterior).

## 4.4.1. Ecuaciones de **Difusión** y **de Onda** (Solución de problemas de valor inicial).
## **Método FTCS: Espacio-Centrado en Tiempo hacia Adelante (<i>Forward-Time Centered-Space</i>)**.

## $\;$

## 4.4.1.1. **Ecuación de Difusión**.

### Ahora abordemos la solución numérica de la Ecuación de Difusión (en dos dimenciones):

<h1 align="center">$ \nabla^2 \phi = \frac{1}{D}\frac{\partial \phi}{\partial t}$</h1>

### que podemos reescribir de la siguiente manera:

<h1 align="center">$ \frac{\partial \phi}{\partial t} = D \nabla^2 \phi$</h1>

### Nótese que en este caso $\phi$, es una función tanto espacial, como temporal, i.e. $\phi(x,y,t)$. Pero, para reolverla procedermos de manera muy similar a los casos anteriores. Primero substituiremos nuestra aproximación numérica del Laplaciano:

<h3 align="center">$ \color{red} \boxed{ \color{black} \nabla^2 \phi = \displaystyle \dfrac{\phi(x+h,y) + \phi(x-h,y)+ \phi(x,y+h) + \phi(x,y-h) - 4 \phi(x,y)}{h^2} }$</h3>

### en la ecuación, por lo que tendremos que:

<h2 align="center">$ \color{blue} \boxed{ \color{black} \displaystyle \dfrac{\partial \phi}{\partial t} = \frac{D}{h^2}  \left[ \phi(x+h,y,t) + \phi(x-h,y,t)+ \phi(x,y+h,t) + \phi(x,y-h,t) - 4 \phi(x,y,t) \right] }$</h2>

<!-- ### Podemos notar que, el lado izquierdo de la ecuación es una función $\phi(x,y)$, que además tiene una dependencia temporal. Por lo que: -->
### Por lo que en general, tenemos que:

<h2 align="center">$\dfrac{\partial \phi}{\partial t} = \displaystyle F \left[ \phi(x,y),t \right]$</h2>

### De esta manera, si conocieramos $\phi(x,y)$ en cada punto, el problema se convierte en una Ecuación Diferencial Ordinaria de primer orden, que podemos resolver con los métodos de Euler o Runge-Kutta, que ya hemos visto.

<h2 align="center">$\dot{\phi} = \displaystyle F \left( \phi,t \right)$</h2>

### Así entonces, podemos combinar el método de Euler o Runge-Kutta, con el _Método de Diferencias Finitas_ (que acabamos de ver), para resolver la ecuación. A esta combinación se le conoce como el método **_Espacio-Centrado en Tiempo hacia Adelante_** o _**FTCS**_ (por sus siglas en inglés).

### El objetivo entonces es usar un método para resolver EDOs (Euler o Runge-Kutta), pero en cada iteración además hay que resolver la malla por método de relajación para la parte temporal, como vimos anteriromente.


### Abordemos la ecuación de Difusón en una dimensión, para poder entender esto mejor. En 1D, la ecuación se reduce a:

<h1 align="center">$ \frac{\partial \phi}{\partial t} = D \frac{\partial^2 \phi}{\partial x^2}$</h1>

### Y al substituir la aproximación numérica para $\frac{\partial^2 \phi}{\partial x^2}$, tenemos que:

<h1 align="center">$ \frac{\partial \phi}{\partial t} = \frac{D}{h^2} \displaystyle \left[ \phi(x+h,t) + \phi(x-h,t) - 2 \phi(x,t) \right]$</h1>

### Por lo que podemos reescribir el problema como: 

<h2 align="center">$\dfrac{\partial \phi}{\partial t} = \displaystyle F \left[ \phi_n(x),t \right]$</h2>

### O bien como un problema de Cauchy:

<h1 align="center">$ \dot{\phi}  = \displaystyle F \left( \phi_n,t \right); \; \; \; F ( \phi_0,t_0 ) = F_0$</h1>

### Donde $F \left( \phi_n,t \right)$, es la solucuón a la Ecuación de Difusión.

### Así entonces tenemos que:

<h2 align="center">$\dot{\phi} = F(\phi,t)$</h2>

### es una ecuación diferencial de 1er orden que podemos resolver mediante el Método de Euler (o Runge-Kutta), considerando sólo la parte temporal de la ecuación:

<h2 align="center">$\Rightarrow \; F(\phi_n, t+\Delta t) = F(\phi_n, t)  + \Delta t \cdot F(\phi_n,t)$</h2>

### Lo anterior, implica que en cada iteración del método de Euler hay que conocer $\phi_{n}$ (la parte espacial), que puede aproximarse mediante el método de relajación iterando desde $\phi_{0}$ hasta que el valor converja. 

<!-- <h1 align="center">Método FTCS: Espacio-Centrado en Tiempo hacia Adelante (<i>Forward-Time Centered-Space</i>).</h1> -->

### Así entonces considerando la parte espacial, tenemos que:

## $\; \; \; \; \; \; F \left[ \phi(x), t + \Delta t \right] = F \left[ \phi(x), t \right]  + \Delta t \cdot F \left[ \phi(x),t \right]$

## $ \Rightarrow \; F \left[ \phi(x), t + \Delta t \right] = F \left[ \phi(x), t \right]  +  \Delta t \cdot \frac{D}{h^2} \displaystyle \left[ \phi(x+h,t) + \phi(x-h,t) - 2 \phi(x,t) \right]$

### Al usar el método de relajación, podemos iterar (con $n$) la parte espacial, por lo que la ecuación puede reescribirse como:

<h2 align="center">$ \color{blue} \boxed{ \color{black} F \left[ \phi_n(x), t + \Delta t \right] = F \left[ \phi_{n-1}(x), t \right]  +  \Delta t \cdot \frac{D}{h^2} \left[ \phi_{n-1}(x+h,t) + \phi_{n-1}(x-h,t)- 2 \phi_{n-1}(x,t) \right]}$</h2>

### donde las condicones a la frontera e iniciales están dadas por:  $\, F [ \phi_0(x),t_0 ] = F_0$

### Ahora bien, para poder abordar el problema de manera más accesible, podemos separar tanto la parte espacial como temporal y usar tanto las condicones a la frontera como las condiciones iniciales. Primero supongamos que hemos resuelto la parte temporal y que conocemos $\phi(x)$ para todo tiempo $t$. Por lo que sólo tendríamos que abordar la parte temporal:

<h2 align="center">$ \color{orange} \boxed{ \color{black} \displaystyle F \left( \phi_n, t + \Delta t \right) = F \left( \phi_n, t \right)  + \Delta t \cdot F \left( \phi_n,t \right)}$</h2>

### Esto es nada más que el método de Euler iterando en cada tiempo $t + \Delta t$. Sin embargo, lo anterior implica que hemos resuelto en cada tiempo $t + \Delta t$, la ecuación espacial $\phi(x)$ (lo cual es posible usando el método de relajación, iterando de $\phi_0$ a $\phi_n$ **sobre todo el espacio $x$** hasta que el valor converja. Sin embargo, podemos hacer esto mismo enntre cada tiempo $t + \Delta t$ (pues es una aproximación númerica), suponiendo que $t$ está fijo (o que no hay dependencia temporal, para ese pequeño lapso de tiempo), lo que es equivalente a tener un tiempo fijo $t_p$.
### Así, tenemos que:
<h2 align="center">$ \phi_n(x,t_p) = \displaystyle \phi_{n-1}(x,t_p) +  D \cdot \frac{\Delta t}{h^2} \left[ \phi_{n-1}(x+h,t_p) + \phi_{n-1}(x-h,t_p)_{n-1} - 2 \phi_{n-1}(x,t_p) \right]$</h2>

### Que de cierta forma es como si tuvieramos una ecuación **solamente temporal** duarante el pequeño perido de tiempo $t + \Delta t$:


<h2 align="center">$ \color{orange} \boxed{ \color{black} \displaystyle  \phi_n(x) = \displaystyle \phi_{n-1}(x) +  D \cdot \frac{\Delta t}{h^2} \left[ \phi_{n-1}(x+h) + \phi_{n-1}(x-h)_{n-1} - 2 \phi_{n-1}(x) \right] }$</h2>

### y que podemos abordar con el método de relajación iterando desde $\phi_0$ hasta que el valor converja ($\phi_n$), para todos los puntos $x$ en el espacio discretizado. 

## * **Notese que el cociente $\dfrac{\Delta t}{h^2}$, contiene tanto el paso de integración temporal ($\Delta t$), como el espacial ($h^2$)**.
#### Por lo que hay que tener cuidado de como definirlos.

### Abordemos un ejemlo con la intensión de aclarar cono funciona el método el método _**FTCS**_ (**_Espacio-Centrado en Tiempo hacia Adelante_**).

## **Ejemplo**: Difsuión de Calor:

### Consideremos la base plana de un recipiente de acero inoxidable de 1 cm de espesor, la cual se encuentra inicialmente a una temperatura uniforme de 20°C. El recipiente se coloca en un baño de agua fría a 0°C y se llena con agua caliente a 50°C. 

<center><img src="img/Heat.png"/></center>

### El objetivo es calcular el perfil de temperatura del acero en función de la distancia $x$ desde el lado caliente hasta el lado frío, en función del tiempo. 

### Para simplificar, consideremos la base del recipiente con un ancho arbitrario, de modo que su perfil de temperatura sea uniforme, y supongamos que ni el agua caliente ni la fría cambian de temperatura apreciablemente. La conducción térmica se rige por la ecuación de difusión (**de calor** en este contexto), por lo que resolver este problema consiste en resolver la ecuación de difusión unidimensional para la temperatura T.

<h1 align="center">$ \frac{\partial T}{\partial t} = D \frac{\partial^2 T}{\partial x^2}$</h1>

## Así entonces, discretizando en tiempo tenemos que:

## $\dfrac{\Delta T}{\Delta t} = D \dfrac{\partial^2 T}{\partial x^2}$

## $\Delta T = \Delta t \cdot D \dfrac{\partial^2 T}{\partial x^2}$

## $ \Rightarrow \; T(x,t+\Delta t) - T(x,t) = \Delta t \cdot D \dfrac{\partial^2 T}{\partial x^2}$

## $ \Rightarrow \; T(x,t+\Delta t) - T(x,t) = \displaystyle \Delta t \cdot \frac{D}{h^2}  \left[ T(x+h,t) + T(x-h,t) - 2 T(x,t) \right]$

## $ \Rightarrow \; T_n(x,t+\Delta t) - T_{n-1}(x,t) = \displaystyle \Delta t \cdot \frac{D}{h^2}  \left[ T_{n-1}(x+h,t) + T_{n-1}(x-h,t) - 2 T_{n-1}(x,t) \right]$

<h1 align="center">$ \color{red} \boxed{ \color{black} \therefore \;\displaystyle T_n(x,t+\Delta t) = T_{n-1}(x,t) + D \cdot \frac{\Delta t}{h^2} \displaystyle \left[ T_{n-1}(x+h,t) + T_{n-1}(x-h,t) - 2 T_{n-1}(x,t) \right] }$</h1>
